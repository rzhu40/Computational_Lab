```matlab
!rm fourier_trans.m
```

```matlab
%%file fourier_trans.m
function [freqs, sqmod] = fourier_trans(x, y, dx_samp)
    idx = mod(x, dx_samp) == 0;
    Y = fft(y(idx));
    f_samp = 1/dx_samp;
    F = fftshift(Y);
    Nq = 1/(2*dx_samp);
    L = x(end);
    freqs = -Nq:1/L:Nq-(1/L);
    sqmod = abs(F).^2;
end
```

```matlab
for i = 1:5
    [f, Fsq] = fourier_trans(X, y, dxs(i));
    subplot(5,1,i)
    plot(f, Fsq)   
end
    xlabel('f (Hz)')
    subplot(5,1,3)
    ylabel('|F|^2')
```

## Q10

```matlab
dx = 0.05;
L = 10;
x = dx:dx:L;
f1 = 0.5;
f2 = 2;
f3 = 5;
y = sin(2*pi*f1*x) + sin(2*pi*f2*x) + sin(2*pi*f3*x);
```

```matlab
g = randn(size(x));
y = g+y;
plot(x, g)
xlabel('x')
ylabel('Noised y signal')
```

The waveform is now extremely corrupted and it is not possible to visually identify the pres- ence of periodic components. By eye, it looks indistinguishable from random chaotic noise

## Q11

```matlab
[freqs, Fsq] = fourier_trans(x, y, 0.05);
plot(freqs, Fsq)
xlabel('f (Hz)')
ylabel('|F+G|^2')
```

Although the waveform is so corrupted, the spectral peaks are still prominent. This demon- strates that Fourier transforms can easily recover periodic signals that have been corrupted by white Gaussian noise, even when the noise levels are significant. They are very useful for identifying periodic signals buried within noise.

## Q12

```matlab
x = 0.2:0.2:6*pi;
s = sawtooth(x);
plot(x, s)
xlabel('x')
ylabel('Sawtooth(x)')
```

```matlab
Y = fft(s);
F = fftshift(Y);
r = abs(F);
theta = angle(F);
Nq = 1/(2*0.2);
L = 6*pi;
freqs = -Nq:1/L:Nq-1/L;
```

```matlab
subplot(2,1,1)
plot(freqs, r)
ylabel('r')

subplot(2,1,2)
plot(freqs, theta)
xlabel('f (Hz)')
ylabel('\theta')
```

```matlab
comb1 = r.*exp(i.*theta);
if1 = ifft(ifftshift(comb1), 'symmetric');
if2 = ifft(ifftshift(r), 'symmetric');
plot(if1, 'DisplayName', 'Original')
hold on
plot(if2, 'DisplayName', 'Phase-modified')
xlabel('x')
ylabel('s')
legend()
```

The phase of a sinusoid encodes its horizontal displacement. Changing the phases of the various sinusoidal components changes their positions relative to each other, and therefore where they constructively/destructively add to produce different features of the waveform. The dominant period is not changed because the relative amplitudes of the sinusoids dont change when the phase is modified, so the frequency component that had the largest amplitude in the original signal will still be the one with the largest amplitude after the phases have been changed.
