## Q5

```matlab
%%file estimate_pi.m
function pi_est = estimate_pi(ns)
    ns = round(ns);
    x = 2*rand(ns,1) - 1;
    y = 2*rand(ns,1) - 1;
    nc = sum(x.^2+y.^2<1);
    pi_est = 4*nc/ns;
end
```

```matlab
n = 10.^[1:0.05:7];
est = zeros(length(n), 1);
for i = 1:1:length(n)
    est(i) = estimate_pi(n(i));
end

semilogx(n, est, '.')
ylim([2.8, 3.5])
ylabel('\pi estimate')
xlabel('Number of iterations')
```

## Q6

```matlab
nit = 1000;
y = zeros(1,nit);
for i = 2:1:nit
    temp = randi([0,1]);
    if temp
        y(i) = y(i-1)+1;
    else
        y(i) = y(i-1)-1;
    end
end

plot(y)
xlabel('Iteration')
ylabel('Position')
```

## Q7

```matlab
%%file walker_1d.m
function end_pos = walker_1d(nsteps)
    y = zeros(1,nsteps);
    for i = 2:1:nsteps
        temp = randi([0,1]);
        if temp
            y(i) = y(i-1)+1;
        else
            y(i) = y(i-1)-1;
        end
    end
    end_pos = y(end);
end
```

```matlab
walker_1d(1000)
```

## Q8

```matlab
samples = zeros(10000,1);
for i = 1:1:10000
    samples(i) = walker_1d(1000);
end

hist(samples, -200:200)
xlabel('Position')
ylabel('Counts')
```

## Q9

```matlab
n_steps = 10:10:1000;
pos_std = zeros(length(n_steps), 1);
for i = 1:1:length(n_steps)
    temp = zeros(10000, 1);
    for j = 1:1:10000
        temp(j) = walker_1d(n_steps(i));
    end
    pos_std(i) = std(temp);
end
```

```matlab
plot(sqrt(n_steps), pos_std)
xlabel('Sqrt(n)')
ylabel('Standard deviation of final position')
```
